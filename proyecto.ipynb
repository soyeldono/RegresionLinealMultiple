```python
import numpy as np
import pandas as pd
```

```python
def read_data(route):
    return pd.read_csv(route)
```

```python
data = read_data("datos_EM.csv")
data
```

```python
X = pd.DataFrame({"x0":[1 for _ in range(data.shape[0])]})
Y = data["y"]
del data["y"]
X = pd.concat([X,data],axis=1)
X
```

## Operaciones

$X^{T}X$

```python
xT_x = np.dot(np.array(X).transpose(),np.array(X))
xT_x
```

$(X^{T}X)^{-1}$

```python
inv_xT_x = np.linalg.inv(xT_x)
inv_xT_x
```

$X^{T}y$

```python
xT_y = np.dot(np.array(X).transpose(),Y)
xT_y
```

$(X^{T}X)^{-1}X^{T}y$ ====> $b$

```python
inv_xT_x_xT_y = np.dot(inv_xT_x,xT_y)
inv_xT_x_xT_y
```

```python
class model:
    
    def __init__(self,params,variables):
        self.params = params
        self.variables = variables
        self.equation = self.__createEquation(params,variables)
        self.model = self.__dictEquation(params,variables)
        
    def __repr__(self):
        return "y = "+self.equation[:-2]
    
    def __getitem__(self,key):
        return self.equation.split("+")[key]
    
    def __createEquation(self,params,variables):        
        modelEquation = ""
        for j,i in enumerate(params):
            modelEquation += str(i) + (variables[j-1] if j != 0 and j-1 < len(variables) else "") + " " + "+ "
        return modelEquation[:-1]
    
    def __dictEquation(self,params,variables):
        modelDict = {}
        for i,j in zip(params[1:],variables):
            modelDict[j] = i
        return modelDict
    
    def predict(self,X):
        """
        X: DataFrame
        
        return: np.array with predictions
        """
        bias = float(self.equation.split("+")[0])
        result = 0
        for i in self.model:
            result += self.model[i]*X[i]
        return np.array(result+bias)
        
```

$y=-73.08+0.87x1+13.97x2+1.2x3+0.58x4+29.7x5$

```python
modelToPredict = model(inv_xT_x_xT_y,list(data.columns))
modelToPredict
```

### Predicciones

```python
yP = modelToPredict.predict(data)
yP
```

```python
u = np.array(Y-yP)
u
```

```python
Sr_2 = np.sum((Y-yP)**2)/(data.shape[0]-(len(data.columns)+1))
Sr_2
```

## Calculando los intervalos de confianza

Grados de libertad = 20

Tstudent_20_0.95 = 1.72

```python
var_bG = Sr_2*inv_xT_x
pd.DataFrame(var_bG)
```

```python
IC_bG_max = []
IC_bG_min = []
IC_bG = []

for i in range(var_bG.shape[0]):
    IC_bG_max.append(inv_xT_x_xT_y[i]+(1.72*((var_bG[i][i])**(1/2))))
    IC_bG_min.append(inv_xT_x_xT_y[i]-(1.72*((var_bG[i][i])**(1/2))))
    IC_bG.append((IC_bG_min[i],IC_bG_max[i]))

IC_bG
```
